```
using System;

public interface IInsurable
{
    void CheckInsuranceStatus();
}

public class CreditCard
{
    private string cardNumber;
    private string holderName;
    private string expiryDate;
    private decimal balance;
    private bool isActive;

    public string CardNumber
    {
        get { return cardNumber; }
        set { cardNumber = value; }
    }

    public string HolderName
    {
        get { return holderName; }
        set { holderName = value; }
    }

    public string ExpiryDate
    {
        get { return expiryDate; }
        set { expiryDate = value; }
    }

    public decimal Balance
    {
        get { return balance; }
        set { balance = value; }
    }

    public bool IsActive
    {
        get { return isActive; }
        set { isActive = value; }
    }

    public CreditCard(string _CardNumber, string _HolderName, string _ExpiryDate, decimal _Balance)
    {
        CardNumber = _CardNumber;
        HolderName = _HolderName;
        ExpiryDate = _ExpiryDate;
        Balance = _Balance;
        IsActive = true;
    }

    public virtual string GetInfo()
    {
        return $"Информация о карте: \nНомер карты: {CardNumber}, Имя холдера: {HolderName}, Срок действия карты: {ExpiryDate}, Активность: {IsActive}";
    }

    public virtual void Pay(decimal sum)
    {
        if (sum > Balance)
        {
            Console.WriteLine("На вашем счете недостаточно средств");
        }
        else if (!IsActive)
        {
            Console.WriteLine("Карта неактивна. Проверьте статус.");
        }
        else
        {
            Balance -= sum;
            Console.WriteLine($"С вашего счета {CardNumber} списано: {sum}. Оставшийся баланс: {Balance}");
        }
    }

    public virtual void CheckBalance()
    {
        Console.WriteLine($"Баланс карты {CardNumber}: {Balance}");
    }

    public void DeactivateCard()
    {
        IsActive = false;
        Console.WriteLine($"Карта {CardNumber} была деактивирована.");
    }

    public void ActivateCard()
    {
        IsActive = true;
        Console.WriteLine($"Карта {CardNumber} активирована.");
    }
}

public class GoldCreditCard : CreditCard, IInsurable
{
    private decimal bonusMiles;
    private decimal insuranceCoverage;

    public decimal BonusMiles
    {
        get { return bonusMiles; }
        set { bonusMiles = value; }
    }

    public decimal InsuranceCoverage
    {
        get { return insuranceCoverage; }
        set { insuranceCoverage = value; }
    }

    public GoldCreditCard(string _CardNumber, string _HolderName, string _ExpiryDate, decimal _Balance, decimal _BonusMiles, decimal _InsuranceCoverage)
        : base(_CardNumber, _HolderName, _ExpiryDate, _Balance)
    {
        BonusMiles = _BonusMiles;
        InsuranceCoverage = _InsuranceCoverage;
    }

    public override void Pay(decimal sum)
    {
        base.Pay(sum);
        int earnedMiles = (int)(sum / 10);
        BonusMiles += earnedMiles;
        Console.WriteLine($"Заработано {earnedMiles} бонусов. Всего бонусов: {BonusMiles}");
    }

    public void CheckInsuranceStatus()
    {
        Console.WriteLine($"Страховое покрытие для карты {CardNumber}: {InsuranceCoverage} единиц.");
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $", Бонусов: {BonusMiles}, Страхование: {InsuranceCoverage}";
    }
}

public class PremiumCreditCard : CreditCard
{
    private string supportAssistant;
    private decimal cashbackRate;

    public string SupportAssistant
    {
        get { return supportAssistant; }
        set { supportAssistant = value; }
    }

    public decimal CashbackRate
    {
        get { return cashbackRate; }
        set { cashbackRate = value; }
    }

    public PremiumCreditCard(string _CardNumber, string _HolderName, string _ExpiryDate, decimal _Balance, string _SupportAssistant, decimal _CashbackRate)
        : base(_CardNumber, _HolderName, _ExpiryDate, _Balance)
    {
        SupportAssistant = _SupportAssistant;
        CashbackRate = _CashbackRate;
    }

    public override void Pay(decimal sum)
    {
        base.Pay(sum);
        decimal cashback = sum * CashbackRate / 100;
        Balance += cashback;
        Console.WriteLine($"Кэшбэк: {cashback}. Баланс после кэшбэка: {Balance}");
    }

    public override void CheckBalance()
    {
        base.CheckBalance();
        if (Balance < 100)
        {
            Console.WriteLine($"Ваш баланс меньше требуемого, обратитесь в поддержку: {SupportAssistant}.");
        }
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $", Тех. Поддержка: {SupportAssistant}, Кэшбэк: {CashbackRate}%";
    }
}

public class CorporateCreditCard : CreditCard, IInsurable
{
    private string company;
    private decimal spendingLimit;

    public string Company
    {
        get { return company; }
        set { company = value; }
    }

    public decimal SpendingLimit
    {
        get { return spendingLimit; }
        set { spendingLimit = value; }
    }

    public CorporateCreditCard(string _CardNumber, string _HolderName, string _ExpiryDate, decimal _Balance, string _Company, decimal _SpendingLimit)
        : base(_CardNumber, _HolderName, _ExpiryDate, _Balance)
    {
        Company = _Company;
        SpendingLimit = _SpendingLimit;
    }

    public override void Pay(decimal sum)
    {
        if (sum > SpendingLimit)
        {
            Console.WriteLine("Превышен лимит расходов для корпоративной карты.");
        }
        else
        {
            base.Pay(sum);
        }
    }

    public void CheckInsuranceStatus()
    {
        Console.WriteLine($"Компания {Company} имеет корпоративное страхование.");
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $", Компания: {Company}, Лимит расходов: {SpendingLimit}";
    }
}

public class Program
{
    public static void Main(string[] args)
    {
        GoldCreditCard goldCard = new GoldCreditCard("1234 5678 9012 3456", "Елена Головач", "31/12/2026", 5000, 100, 20000);
        PremiumCreditCard premiumCard = new PremiumCreditCard("9876 5432 1098 7654", "Максим Комаров", "30/06/2025", 800, "Секретарь банан", 5);
        CorporateCreditCard corporateCard = new CorporateCreditCard("4567 8901 2345 6789", "Иван Иванов", "15/09/2024", 10000, "АйАйАй", 3000);
        GoldCreditCard goldCard = new GoldCreditCard("1234 5678 9123 4567", "Дедка Колобок", "31/12/2026", 5000, 100, 20000);
        PremiumCreditCard premiumCard = new PremiumCreditCard("9876 5432 1098 7654", "Егор Низинчук", "30/06/2025", 800, "Секретарь какой-то", 5);
        CorporateCreditCard corporateCard = new CorporateCreditCard("4567 8901 2345 6789", "Генадий Горин", "15/09/2024", 10000, "ооо Что-то там", 3000);

        Console.WriteLine(goldCard.GetInfo());
        goldCard.Pay(150);
        goldCard.CheckInsuranceStatus();

        Console.WriteLine("\n" + premiumCard.GetInfo());
        premiumCard.CheckBalance();
        premiumCard.Pay(500);

        Console.WriteLine("\n" + corporateCard.GetInfo());
        corporateCard.Pay(2500);
        corporateCard.CheckInsuranceStatus();
    }
}
```
